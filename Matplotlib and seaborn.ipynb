# CATEGORICAL PLOTS

#There are a few main plot types for this:

factorplot,
boxplot,
violinplot,
stripplot,
swarmplot,
barplot,
countplot

```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y= 'total_bill', data = tips)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

Countplot - Counts number of occurence

```python
sns.countplot(x = 'day', data = tips)
```

Box and violinplots


used to show distribution of categorical data. 
A box plot is used for quantitative data analysis, the box shows the quartiles of the dataset, whiskers show  the rest of the distribution, except for the outliers

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
#can do entire dataset with orient = h
sns.boxplot(data = tips, palette = 'rainbow', orient = 'h')
```

```python
sns.boxplot(x='day', y = 'total_bill', data = tips, hue ='sex', palette = 'coolwarm')
```

#violin plot


A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. 

```python
sns.violinplot(x= 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
#swamplot
sns.swarmplot(x='sex', y = 'total_bill', data = tips)
```

Factor plot

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar')
```

# Matrix plot

```python
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
flights.head()
```

```python
tips.head()
```

```python
#inorder for heatmap to work properly we need the data to be in a matrix form
```

```python
tc =tips.corr()
tc
```

```python
sns.heatmap(tc, annot=True)
```

```python
#flights data
flights.head()
```

```python
fc = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')

fc.head()
```

```python
sns.heatmap(fc,cmap = 'magma', lw = 1, linecolor = 'white')
```

#Cluster map

Use hierachial clustering to produce a clustered mapping of the heat map

```python
sns.clustermap(fc)
```

```python
sns.clustermap(fc, cmap = 'coolwarm', standard_scale = 1)
```
